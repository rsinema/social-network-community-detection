```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pathlib import Path
file_path = Path("../data/project_3_graph.graphml")
```

```python
G = nx.read_graphml(file_path)
```

```python
persons = 0
threads = 0
channels = 0

for n in G.nodes:
    node = G.nodes[n]
    # print(node)
    if node.get("node_type") == "person":
        persons += 1
    elif node.get("node_type") == "thread":
        threads += 1
    elif node.get("node_type") == "channel":
        channels += 1

print(persons)
print(threads)
print(channels)
```

```python
pos = nx.spring_layout(G)

def node_color(node_type: str):
    if node_type == "person":
        return "red"
    elif node_type == "thread":
        return "skyblue"
    elif node_type == "channel":
        return "purple"
    else:
        return "yellow"

node_colors = [ node_color(G.nodes[n].get("node_type")) for n in G.nodes]

plt.figure(figsize=(12, 10))
nx.draw(G, pos, node_color=node_colors, node_size=30, edge_color="gray", alpha=0.7, width=0.5)
plt.title("Graph Visualization (spring layout)")
plt.show()
```
